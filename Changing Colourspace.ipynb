There are more than 150 color-space conversion methods available in OpenCV. But we will look into only two, which are most widely used ones: BGR ↔ Gray and BGR ↔ HSV.

For color conversion, we use the function cv.cvtColor(input_image, flag) where flag determines the type of conversion.

For BGR → Gray conversion, we use the flag cv.COLOR_BGR2GRAY. Similarly for BGR → HSV, we use the flag cv.COLOR_BGR2HSV. 

```python
# Load the image using opencv
imgg = cv.imread(r"C:\Users\hp\Desktop\paris.png")
imgg
```

```python
plt.imshow(imgg) # OPENCV uses colour channel BGR(Blue, Green, Red), instead of RGB
plt.title("BGR")
```

```python
# convert from BGR to Gray
imgggray = cv.cvtColor(imgg,cv.COLOR_BGR2GRAY)
plt.imshow(imgggray)
plt.title("BGR to Gray")
```

```python
# convert from BGR to RGB
imggRGB = cv.cvtColor(imgg,cv.COLOR_BGR2RGB)
plt.imshow(imggRGB)
plt.title("BGR to RGB")
```

```python
# convert from BGR to HSV
imggHSV = cv.cvtColor(imgg,cv.COLOR_BGR2HSV)
plt.imshow(imggHSV)
plt.title("BGR to HSV")
```

```python
# Extract blue from object

import numpy as np

cap = cv.VideoCapture(0)
while(1):
    # Take each frame
    _, frame = cap.read()
    # Convert BGR to HSV
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    # define range of blue color in HSV
    lower_blue = np.array([110,50,50])
    upper_blue = np.array([130,255,255])
    # Threshold the HSV image to get only blue colors
    mask = cv.inRange(hsv, lower_blue, upper_blue)
    # Bitwise-AND mask and original image
    res = cv.bitwise_and(frame,frame, mask= mask)
    cv.imshow('frame',frame)
    cv.imshow('mask',mask)
    cv.imshow('res',res)
    if cv.waitKey(1) & 0xff == ord('q'):   # to close the cam
        break
cap.release()
cv.destroyAllWindows()
```

```python
# Extract Green, blue and Red simultaneously using the Captured image

import cv2 as cv
import numpy as np

cap = cv.VideoCapture(0)

while True:
    _, frame = cap.read()
    
    # Convert BGR to HSV
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    # Define range for red color in HSV
    lower_red = np.array([0, 100, 100])
    upper_red = np.array([10, 255, 255])
    
    # Define range for green color in HSV
    lower_green = np.array([40, 50, 50])
    upper_green = np.array([80, 255, 255])
    
    # Define range for blue color in HSV
    lower_blue = np.array([110, 50, 50])
    upper_blue = np.array([130, 255, 255])
    
    # Threshold the HSV image to get only desired colors
    mask_red = cv.inRange(hsv, lower_red, upper_red)
    mask_green = cv.inRange(hsv, lower_green, upper_green)
    mask_blue = cv.inRange(hsv, lower_blue, upper_blue)
    
    # Bitwise-AND masks and original image
    res_red = cv.bitwise_and(frame, frame, mask=mask_red)
    res_green = cv.bitwise_and(frame, frame, mask=mask_green)
    res_blue = cv.bitwise_and(frame, frame, mask=mask_blue)
    
    # Show extracted colors
    cv.imshow('Red', res_red)
    cv.imshow('Green', res_green)
    cv.imshow('Blue', res_blue)
    
    # Show original frame
    cv.imshow('Frame', frame)
    
    if cv.waitKey(1) & 0xff == ord('q'):   # to close the cam
        break
        
cap.release()
cv.destroyAllWindows()
```

```python


# Load an image
image = cv.imread(r"C:\Users\hp\Downloads\parrot.jpg")

# Convert BGR to HSV
hsv = cv.cvtColor(image, cv.COLOR_BGR2HSV)

# Define range of blue color in HSV
lower_blue = np.array([110, 50, 50])
upper_blue = np.array([130, 255, 255])

# Threshold the HSV image to get only blue colors
mask_blue = cv.inRange(hsv, lower_blue, upper_blue)

# Bitwise-AND mask and original image for blue
res_blue = cv.bitwise_and(image, image, mask=mask_blue)

# Show the blue channel
cv.imshow('Blue Channel', res_blue)

# Define range of green color in HSV
lower_green = np.array([50, 100, 100])
upper_green = np.array([70, 255, 255])

# Threshold the HSV image to get only green colors
mask_green = cv.inRange(hsv, lower_green, upper_green)

# Bitwise-AND mask and original image for green
res_green = cv.bitwise_and(image, image, mask=mask_green)

# Show the green channel
cv.imshow('Green Channel', res_green)

# Define range of red color in HSV
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

# Threshold the HSV image to get only red colors (low range)
mask_red1 = cv.inRange(hsv, lower_red, upper_red)

# Define range of red color in HSV (high range)
lower_red = np.array([160, 100, 100])
upper_red = np.array([179, 255, 255])

# Threshold the HSV image to get only red colors (high range)
mask_red2 = cv.inRange(hsv, lower_red, upper_red)

# Combine the masks for red color
mask_red = cv.bitwise_or(mask_red1, mask_red2)

# Bitwise-AND mask and original image for red
res_red = cv.bitwise_and(image, image, mask=mask_red)

# Show the red channel
cv.imshow('Red Channel', res_red)

# Display the original image
cv.imshow('Original Image', image)

# Wait for the user to press any key to close the windows
cv.waitKey(0)
cv.destroyAllWindows()
```

```python

# Load the four images from above
image1 = cv.imread(r"C:\Users\hp\Desktop\original.png")
image2 = cv.imread(r"C:\Users\hp\Desktop\Red.png")
image3 = cv.imread(r"C:\Users\hp\Desktop\green.png")
image4 = cv.imread(r"C:\Users\hp\Desktop\blue.png")

# Resize images to smaller sizes
width, height = 200, 200
image1 = cv.resize(image1, (width, height))
image2 = cv.resize(image2, (width, height))
image3 = cv.resize(image3, (width, height))
image4 = cv.resize(image4, (width, height))

# Create a blank canvas to arrange the images in a grid
canvas = np.zeros((2 * height, 2 * width, 3), dtype=np.uint8)

# Arrange the images in a 2x2 grid
canvas[0:height, 0:width] = image1
canvas[0:height, width:] = image2
canvas[height:, 0:width] = image3
canvas[height:, width:] = image4

# Display the grid of images
cv.imshow('Grid of Images', canvas)

# Wait for the user to press any key to close the window
cv.waitKey(0)
cv.destroyAllWindows()
```
